# Vibration ODE

Solve the vibration ODE,

$$
\begin{gather}
u^{\prime\prime}+\omega^{2}u=0,\nonumber \\
u(0)=I,\ u^{\prime}(0)=0,\ t\in(0,T]
\end{gather}
$$

and plot the results.

```python
import numpy as np
import matplotlib.pyplot as plt

def u_exact(t, I, w):
    return I*np.cos(w*t)

T = 1
Nt = 19
I = 0.25
w = 10

t = np.linspace(0, T, Nt+1)  # mesh points in time
dt = t[1] - t[0]             # constant time step
u = np.zeros(Nt+1)           # solution

u[0] = I
u[1] = u[0] - 0.5*dt**2*w**2*u[0]

for n in range(1, Nt):
    u[n+1] = 2*u[n] - u[n-1] - dt**2*w**2*u[n]

te = np.linspace(0, T,101)
ue = u_exact(te, I, w)

plt.plot(t, u, 'r--o') 
plt.plot(te, ue, 'b-')
plt.xlabel('t')
plt.ylabel('u(t)')
plt.grid()
plt.legend(['numerical', 'exact'], loc='upper right') 
plt.show()
```
